```python
import pandas as pd
import os
```

```python
pwd = os.getcwd()
```

```python
df = pd.read_excel(pwd + '/Data - Survey Monkey Output Edited.xlsx', sheet_name='Edited_Data')
df_mod = df.copy()
```

```python
df_mod.columns
col_drop = ['Start Date', 'End Date', 'Email Address', 'First Name', 'Last Name', 'Custom Data 1']
df_mod = df_mod.drop(columns=col_drop)
df_mod.columns
```

```python
id_vars = list(df_mod.columns)[0:8]
value_vars = list(df_mod.columns)[8:]
df_melt = df_mod.melt(id_vars=id_vars, value_vars=value_vars, var_name='Question + Subquestion', value_name='Answer')
df_melt
```

```python
questions_import = pd.read_excel(pwd + '/Data - Survey Monkey Output Edited.xlsx', sheet_name='Questions')
questions_import
```

```python
questions = questions_import.copy()
questions.drop(columns=['Unnamed: 2'], inplace=True)
questions
```

```python
question = questions_import.copy()
question.drop(columns=['Raw Questions', 'Raw Subquestions','Subquestion','Unnamed: 2'], inplace=True)
question
```

```python
dataset_merged = pd.merge(left=df_melt,right=question,how="left",left_on="Question + Subquestion",right_on="Questions + Sub-Questions")
dataset_merged
```
